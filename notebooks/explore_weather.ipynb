```python
import pandas as pd
import seaborn as sns
```

```python
br_df = pd.read_csv("/Users/sa17/Desktop/projects/projects/weather_api/data/csv/br_final.csv")

br_df = pd.DataFrame(br_df)[["year", "subdivision", "type", "million_60kgs_bag", "nonbear_mill_trees", "bear_mill_trees","avg_unemp_perc"]]

br_df
```

How have harvest outcomes changed for Minas Gerais throughout the years? 

```python
mg_df = br_df[br_df["subdivision"] == "Minas Gerais"]

mg_df
```

```python
sns.lineplot(data=mg_df, x="year", y="million_60kgs_bag", marker="o")
```

Throughout the years, there is a noticeable upward increase in Minas Gerais harvest outcomes. This suggests that harvests in Minas Gerais have become larger over time, with some fluctuations. There was a huge decline around 2004 and 2022 and high peaks in 2003, 2011, 2017 & 2021.

Why have harvest outcomes increase for Minas Gerais?

```python
sns.scatterplot(data=mg_df, x="bear_mill_trees", y="million_60kgs_bag")
```

There is an increase in the harvest outcome in Minas Gerais potentially due to the increase in the number of bearing trees. As the number of bearing trees increased so has the harvest showing a positive correlation.

 Are these harvest outcomes similar to other regions in Brazil? 

```python
sns.lineplot(data=br_df, x="year", y="million_60kgs_bag", hue="subdivision", marker="o")
```

Minas Gerais harvest outcomes are not similar to other regions in Brazil. They have a way better harvest outcome compared to the other regions. Additonally, their harvest numbers is way higher than the rest by more than 10 million. 

Why do Minas Gerais have a better harvest?

```python
sns.boxplot(data=br_df, x="subdivision", y="bear_mill_trees")
```

Minas Gerais having a better harvest than the other regions despite they all having the same amount of bearing trees could be due to them having better weather conditions than the other regions. Better weather conditions makes it easier for them to go out and actually harvest.

Which region produces the most coffee (by sub-category) in Brazil?  

```python
sns.boxplot(data=br_df, x="subdivision", y="million_60kgs_bag", hue="type")
```

Minas Gerais produced the most Arabacia coffee while Espirito Santo produced the most Robusta coffee. Even though Minas Gerais has a better harvest than Espirito Santo, Robusta coffee is only produce in the Espirito Santo region.

How has unemployment changed in the various coffee-producing regions of Brazil? 

```python
sns.lineplot(data=br_df, x="year", y="avg_unemp_perc", hue="subdivision", marker="o")
```

 Each region in Brazil had an increase of unemployment rate from 2014 to 2016, touching their highest peak in 2020 and then rapidly descreasing right after. Parana has consistiently have the lowest unemployment rate throughout the years, followed by Minas Gerais, Espirito Santo and then finally Sao Paulo with the highest unemployment rate.
